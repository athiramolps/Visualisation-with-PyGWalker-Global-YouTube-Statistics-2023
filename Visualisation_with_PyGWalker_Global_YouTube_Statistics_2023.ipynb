# IMPORTING LIBRARIES

```python
import pygwalker as pyg
import pandas as pd
```

# READING DATAFRAME

```python

# Attempt to read the CSV file with different encodings
file_path = '/content/Global YouTube Statistics.csv'

# List of encodings to try
encodings_to_try = ['utf-8', 'ISO-8859-1', 'utf-16']

for encoding in encodings_to_try:
    try:
        df = pd.read_csv(file_path, encoding=encoding)
        # If reading succeeds, break out of the loop
        break
    except UnicodeDecodeError:
        continue

# Now you should be able to read the file with one of the specified encodings
```
